# Python learning objections

## Standard python

* Basic data types
 
    `str`, `int`, `float`
    `list`

* code comments

    `# <your_comment>`

* variables

    `<varible_name> = <value>`

* Assignment operators

    `+=`, `-=`, `*=`, `/+`

* Run a function

    `print()`
    `type()`
    `<varible_name> = <your_function>(<your_input>)`

* define a function

    `def <your_function>():`


* import package/library

    `import pandas as pd`

* for loop

    ```python
    list = [1, 2, 3]
    
    for item in list:
        print(item)

    [output]:
    1
    2
    3
    ```

* if else

    ```python
    if <condition>:
        <code>
    else:
        <code>
    ```

* Logic statement

  * equal to `==`
  * not equal to `!=`
  * greater than `>`
  * less than `<`
  * greater than or equal to `>=`
  * less than or equal to `<=`
  
* Index and slice

    `list[0]`, `list[1:]`
